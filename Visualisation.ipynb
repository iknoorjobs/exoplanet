```
def normal(X):
    Y= (X-np.mean(X))/(np.max(X)-np.min(X))
    return Y
```

```
extrain= extrain.apply(normal,axis=1)
```

```
for i in [0,1,2,3,4]:
    Y = extrain.iloc[i]
    X = np.arange(len(Y)) 
    plt.figure(figsize=(15,5))
    plt.ylabel('Flux')
    plt.xlabel('Observation')
    plt.plot(X, Y)
    plt.show()
```
